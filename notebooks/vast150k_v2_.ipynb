## FULL SETUP

```python
# FULL SETUP
import os, json, subprocess, warnings
import numpy as np
import librosa
import cv2
from tqdm import tqdm
from decord import VideoReader, cpu
```

```python
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# ---- CONFIG ----
SEGMENT_DURATION = 2
FPS_TARGET       = 2
AUDIO_SR         = 16000
MAX_SEGMENTS     = 15000
BATCH_SIZE       = 50

BASE_DIR        = "/content/drive/MyDrive/imagination_dataset"
TMP_VIDEO_DIR   = os.path.join(BASE_DIR, "videos")
OUTPUT_DIR      = os.path.join(BASE_DIR, "segments")   # one subfolder per segment
CHECKPOINT_PATH = os.path.join(BASE_DIR, "checkpoint_v6.json")
ANNOTATION_PATH = os.path.join(BASE_DIR, "VASTannotations150k.json")
```

```python
# ---- SETUP ----

os.makedirs(TMP_VIDEO_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# ---- HELPERS ----
def parse_time_string(t):
    h, m, s = t.split(":")
    return int(h) * 3600 + int(m) * 60 + float(s)

def load_checkpoint():
    if os.path.exists(CHECKPOINT_PATH):
        with open(CHECKPOINT_PATH, "r") as f:
            return json.load(f)
    return {"index": 0, "segments": 0}

def save_checkpoint(state):
    with open(CHECKPOINT_PATH, "w") as f:
        json.dump(state, f)

def download_video(url, output_path):
    try:
        result = subprocess.run(
            ["yt-dlp", "-f", "mp4", "-o" , output_path, url],
            stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL,
            timeout=60
        )
        return (result.returncode == 0
                and os.path.exists(output_path)
                and os.path.getsize(output_path) > 0)
    except Exception:
        return False

def extract_frames(video_path, start):
    """Returns list of raw BGR numpy arrays (H, W, 3), one per target frame."""
    try:
        vr = VideoReader(video_path, ctx=cpu(0))
        fps = vr.get_avg_fps()
        total_frames = len(vr)
        indices = [
            min(int((start + t) * fps), total_frames - 1)
            for t in np.linspace(0, SEGMENT_DURATION, FPS_TARGET, endpoint=False)
        ]
        frames = vr.get_batch(indices).asnumpy()  # RGB, uint8
        return [frames[k] for k in range(len(indices))]
    except Exception:
        return None

def extract_audio(video_path, start):
    try:
        y, _ = librosa.load(video_path, sr=AUDIO_SR, offset=start,
                            duration=SEGMENT_DURATION, mono=True)
        target_len = AUDIO_SR * SEGMENT_DURATION
        if len(y) < target_len:
            y = np.pad(y, (0, target_len - len(y)))
        else:
            y = y[:target_len]
        return y.astype(np.float32)
    except Exception:
        return None

def save_segment(seg_id, frames, audio, text):
    """
    segments/
      seg_000000/
        frame_00.jpg
        frame_01.jpg
        audio.npy
        text.txt
    """
    seg_dir = os.path.join(OUTPUT_DIR, f"seg_{seg_id:06d}")
    os.makedirs(seg_dir, exist_ok=True)

    # Frames — saved as JPEG, full resolution (no resize)
    for k, frame in enumerate(frames):
        frame_path = os.path.join(seg_dir, f"frame_{k:02d}.jpg")
        # decord returns RGB → convert to BGR for cv2
        cv2.imwrite(frame_path, cv2.cvtColor(frame, cv2.COLOR_RGB2BGR),
                    [cv2.IMWRITE_JPEG_QUALITY, 95])

    # Audio — raw float32 waveform
    np.save(os.path.join(seg_dir, "audio.npy"), audio)

    # Text
    with open(os.path.join(seg_dir, "text.txt"), "w", encoding="utf-8") as f:
        f.write(text or "")
```

```python
# ---- LOAD ANNOTATIONS ----
print("Loading annotations...")
with open(ANNOTATION_PATH, "r") as f:
    annotations = json.load(f)
print(f" Setup complete. {len(annotations)} annotations loaded.")
```

```python
# ============================================================
# CELL 2 — RESUME LOOP
# ============================================================

ckpt           = load_checkpoint()
start_idx      = ckpt["index"]
total_segments = ckpt["segments"]
print(f"Checkpoint → index={start_idx}, segments={total_segments}")

# Sync with disk ground truth — count actual seg folders
existing = [d for d in os.listdir(OUTPUT_DIR)
            if os.path.isdir(os.path.join(OUTPUT_DIR, d)) and d.startswith("seg_")]
disk_count = len(existing)

if disk_count != total_segments:
    print(f"⚠️  Mismatch: checkpoint={total_segments}, disk={disk_count} → using disk count")
    total_segments = disk_count

print(f"Resuming from annotation index {start_idx}/{len(annotations)} | segments so far: {total_segments}")

processed_since_ckpt = 0

if not os.path.exists("/content/drive/MyDrive/cookies.txt"):
    raise Exception(" cookies.txt not found — downloads will fail")

for i, ann in enumerate(
        tqdm(annotations[start_idx:], initial=start_idx, total=len(annotations)),
        start=start_idx):

    if total_segments >= MAX_SEGMENTS:
        print(" MAX_SEGMENTS reached.")
        break

    url       = ann.get("url") or ann.get("video_url")
    clip_span = ann.get("clip_span", ["00:00:00.000", "00:00:02.000"])
    text      = ann.get("subtitle") or ann.get("vast_cap") or ""

    if not url:
        continue

    try:
        start = parse_time_string(clip_span[0])
    except Exception:
        continue

    video_path = os.path.join(TMP_VIDEO_DIR, f"vid_{i}.mp4")

    if not download_video(url, video_path):
        continue

    try:
        frames = extract_frames(video_path, start)
        audio  = extract_audio(video_path, start)

        if frames is None or audio is None:
            continue

        save_segment(total_segments, frames, audio, text)
        total_segments += 1
        processed_since_ckpt += 1

    except Exception as e:
        print(f"[{i}] Error: {e}")
    finally:
        if os.path.exists(video_path):
            os.remove(video_path)

    if processed_since_ckpt >= BATCH_SIZE:
        save_checkpoint({"index": i + 1, "segments": total_segments})
        print(f"  checkpoint → index={i + 1}, segments={total_segments}")
        processed_since_ckpt = 0
    import time
    #time.sleep(1.5)  # or higher (2–5s is safer)

save_checkpoint({"index": i + 1, "segments": total_segments})
print(f"\n Done. Total segments: {total_segments}")
```
