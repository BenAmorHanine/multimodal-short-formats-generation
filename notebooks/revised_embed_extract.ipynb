```python
# в”Ђв”Ђ CELL 3: AudioClassifier definition в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import numpy as np

class AudioClassifier:
    """
    Three-class audio segmentation:
      'speech'  в†’ text_trust = 1.0   (Whisper reliable)
      'sound'   в†’ text_trust = 0.5   (audio present, no speech)
      'silence' в†’ text_trust = 0.4   (no signal)
    """
    def __init__(self,
                 speech_energy_threshold=0.015,
                 sound_energy_threshold=0.003,
                 no_speech_threshold=0.6):
        self.speech_energy_threshold = speech_energy_threshold
        self.sound_energy_threshold  = sound_energy_threshold
        self.no_speech_threshold     = no_speech_threshold

    def compute_rms(self, audio_array):
        return float(np.sqrt(np.mean(audio_array.astype(np.float32) ** 2)))

    def get_no_speech_prob(self, whisper_result):
        if not whisper_result:
            return 1.0
        segs = whisper_result.get('segments', [])
        if not segs:
            return 1.0
        return float(np.mean([s.get('no_speech_prob', 1.0) for s in segs]))

    def classify(self, audio_array, whisper_result):
        rms            = self.compute_rms(audio_array)
        no_speech_prob = self.get_no_speech_prob(whisper_result)
        transcript     = (whisper_result or {}).get('text', '').strip()
        has_text       = len(transcript) > 0

        if rms < self.sound_energy_threshold:
            audio_type, text_trust = 'silence', 0.4
        elif has_text and no_speech_prob < self.no_speech_threshold:
            audio_type, text_trust = 'speech', 1.0
        elif rms >= self.speech_energy_threshold and no_speech_prob < 0.8 and has_text:
            audio_type, text_trust = 'speech', 0.75
        else:
            audio_type, text_trust = 'sound', 0.5

        meta = {
            'rms_energy':     rms,
            'no_speech_prob': no_speech_prob,
            'has_transcript': has_text,
        }
        return audio_type, text_trust, meta

audio_classifier = AudioClassifier()
print("вњ“ AudioClassifier ready")
```

```python
# в”Ђв”Ђ CELL 4: Load Whisper + BLIP-1 large + Whisper-AT в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import whisper
from transformers import BlipProcessor, BlipForConditionalGeneration
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

whisper_model = whisper.load_model("base")
print("вњ“ Whisper loaded")

processor  = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
blip_model = BlipForConditionalGeneration.from_pretrained(
    "Salesforce/blip-image-captioning-large",
    torch_dtype=torch.float16
).to(device).eval()
print("вњ“ BLIP-1 large loaded (~1GB VRAM)")

import subprocess
subprocess.run(["pip", "install", "--no-deps", "whisper-at"], check=True)

import whisper_at as whisper_at_model
from whisper_at import parse_at_label

wat_model = whisper_at_model.load_model("base")
print("вњ“ Whisper-AT loaded (ASR + AudioSet tagging)")
```

```python
# в”Ђв”Ђ CELL 5: Segment helpers  + agressive whisperв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import cv2, tempfile, os
import librosa, soundfile as sf
from PIL import Image

def load_audio_segment(video_path, start, end, sr=16000):
    y, _ = librosa.load(video_path, sr=sr, offset=start, duration=(end - start))
    return y

def extract_middle_frame(video_path, start, end):
    cap = cv2.VideoCapture(video_path)
    cap.set(cv2.CAP_PROP_POS_MSEC, ((start + end) / 2) * 1000)
    ret, frame = cap.read()
    cap.release()
    if not ret:
        return None
    return Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

def aggressive_transcribe(audio_path, model):
    """Two-pass Whisper: standard в†’ relaxed if empty."""
    result = model.transcribe(audio_path, language=None, task="transcribe")
    if result['text'].strip():
        return result
    return model.transcribe(
        audio_path,
        no_speech_threshold=0.3,
        logprob_threshold=None,
        compression_ratio_threshold=None,
        condition_on_previous_text=False,
        language=None
    )

def blip1_caption(frame):
    if frame is None:
        return ""
    inputs = processor(frame, return_tensors="pt").to(device, torch.float16)
    with torch.no_grad():
        ids = blip_model.generate(**inputs, max_new_tokens=30)
    return processor.decode(ids[0], skip_special_tokens=True)

def get_video_segments(video_path, window_size=2, stride=1):
    """Build (start, end) time pairs from video duration."""
    cap = cv2.VideoCapture(video_path)
    fps   = cap.get(cv2.CAP_PROP_FPS)
    total = cap.get(cv2.CAP_PROP_FRAME_COUNT)
    cap.release()
    duration = total / fps
    segments = []
    t = 0.0
    while t + window_size <= duration:
        segments.append((round(t, 3), round(t + window_size, 3)))
        t += stride
    return segments

print("вњ“ Helpers ready")
```

```python
#whisper at helpers
#did only use the labelling fct
def whisper_at_transcribe(audio_path, model):
    """
    Single Whisper-AT call: returns transcript + AudioSet logits.
    Two-pass fallback preserved: if transcript is empty, relax thresholds.
    at_time_res=2.0 matches our 2-second window exactly.
    """
    import warnings
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")   # suppress time-res mismatch warning
        result = model.transcribe(audio_path, at_time_res=2.0, language=None)

    if result['text'].strip():
        return result

    # Relaxed second pass (same as before, now via Whisper-AT)
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        result = model.transcribe(
            audio_path,
            at_time_res=2.0,
            no_speech_threshold=0.3,
            logprob_threshold=None,
            compression_ratio_threshold=None,
            condition_on_previous_text=False,
            language=None
        )
    return result


def get_at_label(wat_result, top_k=1, score_threshold=0.0):
    """
    Extract the top AudioSet label(s) from a Whisper-AT result.
    Returns a string like "Applause" or "Music, Piano" (top_k=2).
    Returns "" if no label has a positive score.
    Only uses segment 0 (our window is already 2s, one segment).
    """
    try:
        tags = parse_at_label(wat_result, top_k=top_k,
                               p_threshold=score_threshold)
        # tags is a list of dicts, one per 2s chunk
        if not tags:
            return ""
        audio_tags = tags[0].get('audio tags', [])
        # Filter to positive scores only (confidence > threshold)
        positive = [(name, score) for name, score in audio_tags
                    if score > score_threshold]
        if not positive:
            return ""
        return ", ".join(name for name, _ in positive[:top_k])
    except Exception:
        return ""
```

```python
# в”Ђв”Ђ CELL 6: Stage 2+3 вЂ” Whisper + BLIP-1 + Whipser-AT в†’ segment_texts.json в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import json
from tqdm import tqdm

WINDOW_SIZE = 2
STRIDE      = 1

PROJECT_ROOT   = "/content/drive/MyDrive/PFA"
IMAGEBIND_ROOT = "/content/drive/MyDrive/PFA/ImageBind"
LIB_ROOT       = "/content/drive/MyDrive/PFA/ImageBind/ImageBind"

INPUT_DIR  = os.path.join(PROJECT_ROOT, "data")
OUTPUT_DIR = os.path.join(IMAGEBIND_ROOT, "results")

TARGET_VIDEO = "Obama_Yes_we_can.mp4"
VIDEO_PATH   = os.path.join(INPUT_DIR, TARGET_VIDEO)
VIDEO_STEM   = os.path.splitext(TARGET_VIDEO)[0]




segments = get_video_segments(VIDEO_PATH, WINDOW_SIZE, STRIDE)
print(f"Total segments: {len(segments)}")

segment_data = {}   # { str(i): {text, trust, source, start, end} }

for i, (start, end) in tqdm(enumerate(segments), total=len(segments)):
    try:
        # вЂ” Whisper вЂ”
        audio = load_audio_segment(VIDEO_PATH, start, end)
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as f:
            sf.write(f.name, audio, 16000)
            whisper_result = aggressive_transcribe(f.name, whisper_model)
            #os.unlink(f.name)

        transcript = whisper_result['text'].strip()

        # в”Ђв”Ђ Whisper-AT: transcript + AudioSet tags in one call в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
        wav_path = f.name
        wat_result = whisper_at_transcribe(wav_path, wat_model)
        os.unlink(wav_path)

        # вЂ” Classify вЂ”
        audio_type, text_trust, _ = audio_classifier.classify(audio, whisper_result)

        # вЂ” Text decision: always real text, never zeros вЂ”
        if audio_type == "speech" and transcript:
            combined_text = transcript
            source = "speech"


        #added whisper-at usecase!
        elif audio_type == "sound":
            # Sound: AudioSet label tells WHAT is happening acoustically,
            # BLIP tells WHAT it looks like вЂ” concatenate both.
            at_label = get_at_label(wat_result, top_k=2, score_threshold=0.0)
            frame    = extract_middle_frame(VIDEO_PATH, start, end)
            caption  = blip1_caption(frame)

            parts = [p for p in [at_label, caption] if p]
            combined_text = ". ".join(parts) if parts else "audio scene"
            source = "whisper_at+blip1"


        else:
            frame   = extract_middle_frame(VIDEO_PATH, start, end)
            caption = blip1_caption(frame)
            combined_text = (transcript + " " + caption).strip() or "visual scene"
            source = "blip1" if caption else "fallback"

        segment_data[str(i)] = {
            "text":  combined_text,
            "trust": round(text_trust, 4),
            "source": source,
            "start": start,
            "end":   end
        }

    except Exception as e:
        print(f"Segment {i} error: {e}")
        segment_data[str(i)] = {
            "text": "visual scene", "trust": 0.4,
            "source": "error", "start": start, "end": end
        }

with open(TEXTS_PATH, "w") as f:
    json.dump(segment_data, f, indent=2)

print(f"вњ“ Saved {len(segment_data)} segments в†’ {TEXTS_PATH}")
sources = {}
for v in segment_data.values():
    sources[v['source']] = sources.get(v['source'], 0) + 1
print(f"  Sources breakdown: {sources}")
```

```python
# в”Ђв”Ђ CELL 8: Load ImageBind в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ

from imagebind.models import imagebind_model
from imagebind.models.imagebind_model import ModalityType
from imagebind import data as ib_data

device = "cuda" if torch.cuda.is_available() else "cpu"

imagebind_mod = imagebind_model.imagebind_huge(pretrained=True)
imagebind_mod.eval().to(device)

print("вњ“ ImageBind loaded")
print(f"VRAM after ImageBind: {torch.cuda.memory_allocated()/1e9:.2f} GB")
```

```python
# в”Ђв”Ђ CELL 9 : Stage 1 вЂ” Single ImageBind pass в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import json, numpy as np, torch, torch.nn.functional as F
import tempfile, os
from tqdm import tqdm
from imagebind import data as ib_data
from imagebind.models.imagebind_model import ModalityType

# в†ђ your existing, tested functions
from shared_utils.video_processing import save_video_segment
from shared_utils.audio_processing import extract_audio_segment
import soundfile as sf

with open(TEXTS_PATH) as f:
    segment_data = json.load(f)

N = len(segment_data)
print(f"Processing {N} segments")

vision_embs, audio_embs, text_embs = [], [], []
times_out, raw_texts, trust_vals = [], [], []
errors = []

for i in range(N):
    seg = segment_data[str(i)]
    start, end = seg['start'], seg['end']
    text = seg['text']
    print(f"Segment {i}: {start}в†’{end}  text='{text[:30]}'")

    clip_path  = None
    audio_path = None

    try:
        with tempfile.TemporaryDirectory() as tmpdir:

            # вЂ” Vision: mp4 clip via your save_video_segment вЂ”
            clip_path = os.path.join(tmpdir, "segment.mp4")
            save_video_segment(VIDEO_PATH, start, end, clip_path)

            # вЂ” Audio: wav array via your extract_audio_segment вЂ”
            audio_arr = extract_audio_segment(VIDEO_PATH, start, end, sr=16000)
            audio_path = os.path.join(tmpdir, "segment.wav")
            sf.write(audio_path, audio_arr, 16000)

            # вЂ” ImageBind inputs вЂ”
            inputs = {
                ModalityType.VISION: ib_data.load_and_transform_video_data(
                    [clip_path], device,
                    clip_duration=2,
                    clips_per_video=1
                ),
                ModalityType.AUDIO: ib_data.load_and_transform_audio_data(
                    [audio_path], device
                ),
                ModalityType.TEXT: ib_data.load_and_transform_text(
                    [text], device
                ),
            }

            with torch.no_grad():
                embs = imagebind_mod(inputs)

            vision_embs.append(embs[ModalityType.VISION].squeeze(0).cpu())
            audio_embs.append(embs[ModalityType.AUDIO].squeeze(0).cpu())
            text_embs.append(embs[ModalityType.TEXT].squeeze(0).cpu())

    except Exception as e:
        print(f"Segment {i} error: {e}")
        errors.append(i)
        vision_embs.append(torch.zeros(1024))
        audio_embs.append(torch.zeros(1024))
        text_embs.append(torch.zeros(1024))

    times_out.append([start, end])
    raw_texts.append(text)
    trust_vals.append(seg['trust'])

# вЂ” Stack & normalize вЂ”
V = F.normalize(torch.stack(vision_embs), dim=-1)
A = F.normalize(torch.stack(audio_embs),  dim=-1)
T = F.normalize(torch.stack(text_embs),   dim=-1)
times_arr = np.array(times_out)
trust_arr = np.array(trust_vals, dtype=np.float32)

np.savez(EMB_PATH,
         vision=V.numpy(), audio=A.numpy(), text=T.numpy(),
         raw_text=raw_texts, times=times_arr, text_trust=trust_arr)

print(f"\nвњ“ Saved в†’ {EMB_PATH}")
print(f"  Shapes  вЂ” V:{V.shape}  A:{A.shape}  T:{T.shape}")
print(f"  Errors  вЂ” {len(errors)} segments: {errors if errors else 'none'}")
print(f"\n  Vision non-zero: {V.abs().sum().item():.2f}")
print(f"  Audio  non-zero: {A.abs().sum().item():.2f}")
print(f"  Text   non-zero: {T.abs().sum().item():.2f}")
```

```python
# в”Ђв”Ђ CELL 10: Inspect Final Embeddings в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import numpy as np, torch

npz    = np.load(EMB_PATH, allow_pickle=True)
V      = torch.tensor(npz['vision'])
A      = torch.tensor(npz['audio'])
T      = torch.tensor(npz['text'])
texts  = npz['raw_text']
trust  = npz['text_trust']

print(f"Shapes вЂ” V:{V.shape}  A:{A.shape}  T:{T.shape}")
print(f"\nCross-modal cosine similarities:")
print(f"  Vision-Audio : {(V*A).sum(-1).mean():.3f} В± {(V*A).sum(-1).std():.3f}")
print(f"  Vision-Text  : {(V*T).sum(-1).mean():.3f} В± {(V*T).sum(-1).std():.3f}")
print(f"  Audio-Text   : {(A*T).sum(-1).mean():.3f} В± {(A*T).sum(-1).std():.3f}")

print(f"\nText trust stats: mean={trust.mean():.3f}  min={trust.min():.3f}")
print(f"\nSample texts (with trust):")
for i in [0, 1, 2, len(texts)//2, -1]:
    print(f"  [{i:>3}] trust={trust[i]:.2f}  {texts[i][:80]}")
```

```python
# в”Ђв”Ђ CELL 11: Stage 4 вЂ” Confidence Gate Fusion в”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђв”Ђ
import torch, torch.nn as nn, torch.nn.functional as F

class ConfidenceGate(nn.Module):
    def __init__(self, input_dim=1024, proj_dim=512):
        super().__init__()
        def gate(d):
            return nn.Sequential(nn.Linear(d,256), nn.ReLU(), nn.Linear(256,1), nn.Sigmoid())
        def proj(d, o):
            return nn.Sequential(nn.Linear(d,o), nn.GELU(), nn.LayerNorm(o))
        self.w_v = gate(input_dim * 2)
        self.w_a = gate(input_dim * 2)
        self.w_t = gate(input_dim * 3)
        self.pv  = proj(input_dim, proj_dim)
        self.pa  = proj(input_dim, proj_dim)
        self.pt  = proj(input_dim, proj_dim)

    def forward(self, v, a, t, trust):
        wv = self.w_v(torch.cat([v, a], dim=-1))
        wa = self.w_a(torch.cat([a, v], dim=-1))
        wt = self.w_t(torch.cat([v, a, t], dim=-1)) * trust.unsqueeze(-1)
        fused = (wv*self.pv(v) + wa*self.pa(a) + wt*self.pt(t)) / (wv+wa+wt+1e-8)
        return F.normalize(fused, dim=-1), {
            'vision': wv.squeeze(-1),
            'audio':  wa.squeeze(-1),
            'text':   wt.squeeze(-1)
        }

gate = ConfidenceGate(input_dim=1024, proj_dim=512).eval()

# text_trust comes from the classifier вЂ” NOT all 1.0 anymore
trust_tensor = torch.tensor(trust, dtype=torch.float32)

with torch.no_grad():
    unified, weights = gate(V, A, T, trust_tensor)

print(f"Unified: {unified.shape}  norm={unified.norm(dim=-1).mean():.4f}")
for name, w in weights.items():
    print(f"  {name:<8}: mean={w.mean():.3f} В± {w.std():.3f}")

print("\nвњ“ Stage 4 complete вЂ” unified_embs ready for highlight detection")
```
