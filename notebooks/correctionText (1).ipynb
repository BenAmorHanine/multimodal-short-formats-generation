```python
# CELL 6: Video Processing Functions (with all imports inside functions)

def segment_video(video_path, window_size=2, stride=1):
    import numpy as np
    from decord import VideoReader, cpu

    vr = VideoReader(video_path, ctx=cpu(0))
    fps = vr.get_avg_fps()
    duration = len(vr) / fps

    segments = []
    t = 0
    while t + window_size <= duration:
        segments.append((t, t + window_size))
        t += stride

    print(f"Video duration: {duration:.1f}s, Segments: {len(segments)}")
    return segments, fps, vr

def extract_frames(vr, start_time, end_time, fps, num_frames=8):
    import numpy as np

    start_frame = int(start_time * fps)
    end_frame = int(end_time * fps)
    frame_indices = np.linspace(start_frame, end_frame - 1, num_frames, dtype=int)
    return vr.get_batch(frame_indices).asnumpy()

def extract_audio_segment(video_path, start_time, end_time, sr=16000):
    import os
    import subprocess
    import librosa

    audio_file = video_path.replace('.mp4', '.wav')
    if not os.path.exists(audio_file):
        cmd = ['ffmpeg', '-i', video_path, '-vn', '-acodec', 'pcm_s16le',
               '-ar', str(sr), '-ac', '1', audio_file, '-y']
        subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    audio, _ = librosa.load(audio_file, sr=sr, offset=start_time, duration=end_time-start_time)
    return audio

def extract_transcript_with_timestamps(video_path, whisper_model):
    result = whisper_model.transcribe(video_path, word_timestamps=True, verbose=False)
    segments = []
    for segment in result['segments']:
        segments.append({
            'start': segment['start'],
            'end': segment['end'],
            'text': segment['text'].strip()
        })
    return segments, result['text']

def align_text_to_segments(transcript_segments, video_segments):
    aligned_texts = []
    for vid_path, vid_start, vid_end in video_segments:
        overlapping_text = []
        for trans_seg in transcript_segments:
            if not (trans_seg['end'] < vid_start or trans_seg['start'] > vid_end):
                overlapping_text.append(trans_seg['text'])
        segment_text = " ".join(overlapping_text).strip()
        aligned_texts.append(segment_text)
    return aligned_texts

print("✓ Video processing functions loaded with proper imports")
```

```python
# CELL 7: IF-MMIN Module Definition (FIXED dimensions)

# CELL 7: IF-MMIN Module Definition (COMPLETELY FIXED)

import torch
import torch.nn as nn
import torch.nn.functional as F
from imagebind.models.imagebind_model import ModalityType

class CMDLoss(nn.Module):
    """Central Moment Discrepancy loss"""
    def __init__(self, max_order=5):
        super().__init__()
        self.max_order = max_order

    def forward(self, H_a, H_v, H_t):
        loss = 0.0
        pairs = [('a', 'v'), ('a', 't'), ('v', 't')]

        for m1, m2 in pairs:
            H1 = eval(f'H_{m1}')
            H2 = eval(f'H_{m2}')

            mean_loss = torch.norm(H1.mean(dim=0) - H2.mean(dim=0), p=2)
            loss += mean_loss

            for k in range(2, self.max_order + 1):
                H1_centered = H1 - H1.mean(dim=0, keepdim=True)
                H2_centered = H2 - H2.mean(dim=0, keepdim=True)
                moment1 = torch.mean(H1_centered ** k, dim=0)
                moment2 = torch.mean(H2_centered ** k, dim=0)
                loss += torch.norm(moment1 - moment2, p=2)

        return loss / len(pairs)

class IFMMINModule(nn.Module):
    def __init__(self, embed_dim=1024, specific_dim=256, invariant_dim=384):
        super().__init__()

        print(f"Initializing IF-MMIN with embed_dim={embed_dim}, specific_dim={specific_dim}, invariant_dim={invariant_dim}")

        # Specificity Encoders
        self.specificity_encoders = nn.ModuleDict({
            ModalityType.VISION: nn.Sequential(
                nn.Linear(embed_dim, 512),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(512, specific_dim)
            ),
            ModalityType.AUDIO: nn.Sequential(
                nn.Linear(embed_dim, 512),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(512, specific_dim)
            ),
            ModalityType.TEXT: nn.Sequential(
                nn.Linear(embed_dim, 512),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(512, specific_dim)
            )
        })

        # Invariance Encoder (takes concatenated specific features)
        self.invariance_encoder = nn.Sequential(
            nn.Linear(specific_dim * 3, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, invariant_dim)
        )

        # Split invariant features back to modality-specific
        self.invariance_split = nn.Linear(invariant_dim, specific_dim * 3)

        # IF-IM: 5 cascaded autoencoders
        # FIXED: Input dimension should be (invariant_dim + specific_dim * 2)
        # But we need to ensure the dimensions match!
        input_dim = invariant_dim + specific_dim * 2  # 384 + 512 = 896
        hidden_dim = 512
        output_dim = specific_dim  # 256

        print(f"  IF-IM input_dim: {input_dim}, hidden_dim: {hidden_dim}, output_dim: {output_dim}")

        self.if_im = nn.ModuleList([
            nn.Sequential(
                nn.Linear(input_dim if i == 0 else output_dim + invariant_dim, hidden_dim),
                nn.ReLU(),
                nn.Dropout(0.2),
                nn.Linear(hidden_dim, hidden_dim // 2),
                nn.ReLU(),
                nn.Linear(hidden_dim // 2, output_dim)
            ) for i in range(5)
        ])

        # Project back to ImageBind space
        self.to_imagebind = nn.Sequential(
            nn.Linear(specific_dim, 512),
            nn.ReLU(),
            nn.Linear(512, embed_dim)
        )

        # Initialize weights
        self._init_weights()

        print(f"✓ IF-MMIN initialized with:")
        print(f"  - Specificity encoders: {embed_dim} -> {specific_dim}")
        print(f"  - Invariance encoder: {specific_dim*3} -> {invariant_dim}")
        print(f"  - IF-IM: 5 autoencoders with dynamic dimensions")

    def _init_weights(self):
        for module in self.modules():
            if isinstance(module, nn.Linear):
                nn.init.xavier_uniform_(module.weight)
                if module.bias is not None:
                    nn.init.constant_(module.bias, 0)

    def forward_with_text(self, vision_emb, audio_emb, text_emb):
        # Print shapes for debugging (first iteration only)
        if not hasattr(self, '_debug_printed'):
            print(f"\n[DEBUG] forward_with_text input shapes:")
            print(f"  vision_emb: {vision_emb.shape}")
            print(f"  audio_emb: {audio_emb.shape}")
            print(f"  text_emb: {text_emb.shape}")

        # Get modality-specific features
        h_v = self.specificity_encoders[ModalityType.VISION](vision_emb)
        h_a = self.specificity_encoders[ModalityType.AUDIO](audio_emb)
        h_t = self.specificity_encoders[ModalityType.TEXT](text_emb)

        if not hasattr(self, '_debug_printed'):
            print(f"  h_v: {h_v.shape}")
            print(f"  h_a: {h_a.shape}")
            print(f"  h_t: {h_t.shape}")

        # Get invariant features
        h_concat = torch.cat([h_v, h_a, h_t], dim=1)
        H = self.invariance_encoder(h_concat)

        if not hasattr(self, '_debug_printed'):
            print(f"  h_concat: {h_concat.shape}")
            print(f"  H: {H.shape}")

        # Imagine text from vision+audio
        imagined_text = self._imagine_text(h_v, h_a, H)

        if not hasattr(self, '_debug_printed'):
            print(f"  imagined_text: {imagined_text.shape}")
            self._debug_printed = True

        return {
            'h_t_real': h_t,
            'h_t_imagined': imagined_text,
            'H': H
        }

    def forward_without_text(self, vision_emb, audio_emb):
        h_v = self.specificity_encoders[ModalityType.VISION](vision_emb)
        h_a = self.specificity_encoders[ModalityType.AUDIO](audio_emb)

        # For missing text, we use a learned prior (zeros work with proper training)
        batch_size = h_v.shape[0]
        h_t_prior = torch.zeros(batch_size, h_v.shape[1]).to(h_v.device)

        h_concat = torch.cat([h_v, h_a, h_t_prior], dim=1)
        H = self.invariance_encoder(h_concat)

        imagined_text_specific = self._imagine_text(h_v, h_a, H)
        return self.to_imagebind(imagined_text_specific)

    def _imagine_text(self, h_v, h_a, H):
        """FIXED: Proper dimension handling in cascade"""
        batch_size = h_v.shape[0]

        # For first autoencoder: input = [H, h_v, h_a] concatenated
        h_av = torch.cat([h_v, h_a], dim=1)  # [batch, 512]
        input_feat = torch.cat([H, h_av], dim=1)  # [batch, 384+512=896]

        if not hasattr(self, '_debug_printed_im'):
            print(f"\n[DEBUG] _imagine_text shapes:")
            print(f"  h_v: {h_v.shape}")
            print(f"  h_a: {h_a.shape}")
            print(f"  H: {H.shape}")
            print(f"  h_av: {h_av.shape}")
            print(f"  input_feat (first AE): {input_feat.shape}")
            self._debug_printed_im = True

        # First autoencoder
        delta_z = self.if_im[0](input_feat)  # [batch, 256]

        # Subsequent autoencoders: input = [H, previous output]
        for i in range(1, len(self.if_im)):
            input_feat = torch.cat([H, delta_z], dim=1)  # [batch, 384+256=640]
            if not hasattr(self, '_debug_printed_im'):
                print(f"  input_feat (AE {i}): {input_feat.shape}")
            delta_z = self.if_im[i](input_feat)

        if not hasattr(self, '_debug_printed_im'):
            print(f"  final delta_z: {delta_z.shape}")
            self._debug_printed_im = True

        return delta_z

class IFMMINTrainer:
    def __init__(self, device):
        self.device = device
        self.if_mmin = IFMMINModule().to(device)
        self.cmd_loss = CMDLoss()
        self.optimizer = torch.optim.Adam(self.if_mmin.parameters(), lr=0.0002)
        print("✓ IF-MMIN Trainer initialized")

    def training_step(self, vision_emb, audio_emb, text_emb):
        self.optimizer.zero_grad()
        outputs = self.if_mmin.forward_with_text(vision_emb, audio_emb, text_emb)

        # Split H into H_v, H_a, H_t for CMD loss
        H_split = self.if_mmin.invariance_split(outputs['H'])
        H_v, H_a, H_t = H_split.chunk(3, dim=1)

        loss_cmd = self.cmd_loss(H_v, H_a, H_t)
        loss_img = F.mse_loss(outputs['h_t_imagined'], outputs['h_t_real'])

        total_loss = loss_cmd + loss_img

        total_loss.backward()
        self.optimizer.step()

        return {
            'total_loss': total_loss.item(),
            'cmd_loss': loss_cmd.item(),
            'img_loss': loss_img.item()
        }

    def imagine_text(self, vision_emb, audio_emb):
        with torch.no_grad():
            return self.if_mmin.forward_without_text(vision_emb, audio_emb)

    def save(self, path):
        torch.save(self.if_mmin.state_dict(), path)
        print(f"✓ Model saved to {path}")

    def load(self, path):
        self.if_mmin.load_state_dict(torch.load(path, map_location=self.device))
        print(f"✓ Model loaded from {path}")

print("✓ IF-MMIN module loaded successfully with FIXED dimensions!")
```

```python
# CELL 8: Enhanced Processing with IF-MMIN (with all imports)

# CELL 8: Enhanced Processing with IF-MMIN (FIXED - store averaged vision)

def process_segment_with_ifmmin(vr, video_path, start_time, end_time, fps,
                                segment_text, ifmmin_trainer=None, num_frames=8):
    """Process segment with intelligent text handling using IF-MMIN"""
    import os
    import tempfile
    import numpy as np
    import torch
    import soundfile as sf
    from PIL import Image
    from imagebind import data
    from imagebind.models.imagebind_model import ModalityType

    # Extract frames
    frames = extract_frames(vr, start_time, end_time, fps, num_frames)

    with tempfile.TemporaryDirectory() as tmpdir:
        # Save frames
        frame_paths = []
        for i, frame in enumerate(frames):
            frame_path = os.path.join(tmpdir, f"frame_{i}.jpg")
            Image.fromarray(frame).save(frame_path)
            frame_paths.append(frame_path)

        # Extract and save audio
        audio = extract_audio_segment(video_path, start_time, end_time)
        audio_path = os.path.join(tmpdir, "audio.wav")
        sf.write(audio_path, audio, 16000)

        # Get vision and audio embeddings from ImageBind
        inputs = {
            ModalityType.VISION: data.load_and_transform_vision_data(frame_paths, device),
            ModalityType.AUDIO: data.load_and_transform_audio_data([audio_path], device)
        }

        with torch.no_grad():
            embeddings = model(inputs)

        # Vision: [num_frames, 1024] - keep all frames for flexibility
        vision_emb = embeddings[ModalityType.VISION].cpu().numpy()
        # Audio: [1, 1024]
        audio_emb = embeddings[ModalityType.AUDIO].cpu().numpy()

        # Handle text intelligently
        if segment_text and len(segment_text.strip()) > 0:
            # Real text available
            text_inputs = {ModalityType.TEXT: data.load_and_transform_text([segment_text], device)}
            with torch.no_grad():
                text_emb = model(text_inputs)[ModalityType.TEXT].cpu().numpy()
            text_source = "real"
            text_weight = 1.0

            # For combined embedding, use averaged vision
            vision_avg = vision_emb.mean(axis=0, keepdims=True)

        else:
            # No text - use IF-MMIN to imagine it
            if ifmmin_trainer is not None:
                # Use averaged vision for imagination
                vision_avg = vision_emb.mean(axis=0, keepdims=True)
                vision_tensor = torch.from_numpy(vision_avg).float().to(device)
                audio_tensor = torch.from_numpy(audio_emb).float().to(device)

                imagined_text = ifmmin_trainer.imagine_text(vision_tensor, audio_tensor)
                text_emb = imagined_text.cpu().numpy()
                text_source = "imagined"
                text_weight = 0.8
            else:
                # Fallback to zeros
                vision_avg = vision_emb.mean(axis=0, keepdims=True)
                text_emb = np.zeros_like(vision_avg)
                text_source = "zero_fallback"
                text_weight = 0.0

        # Weighted combination using averaged vision
        total_weight = 1.0 + 1.0 + text_weight
        combined_emb = (vision_avg + audio_emb + text_weight * text_emb) / total_weight

    return {
        'vision': vision_emb,           # Keep all frames [8, 1024]
        'vision_avg': vision_avg,       # Averaged version [1, 1024]
        'audio': audio_emb,              # [1, 1024]
        'text': text_emb,                # [1, 1024]
        'combined': combined_emb,        # [1, 1024]
        'text_source': text_source,
        'start': start_time,
        'end': end_time,
        'original_text': segment_text if segment_text else ""
    }

print("✓ Enhanced processing with IF-MMIN ready (with proper shapes)")
```

```python
# CELL 10: Train IF-MMIN (with fixed paths)

import torch
import numpy as np
import os

# Define results directory if not already defined
if 'results_dir' not in locals():
    results_dir = "/content/drive/MyDrive/PFA/ImageBind/results"
    os.makedirs(results_dir, exist_ok=True)
    print(f"✓ Set results_dir to: {results_dir}")

# Check if we have training data
if 'training_results' in locals() and len(training_results) > 0:

    # Filter segments with real text for training
    text_segments = [r for r in training_results if r['text_source'] == 'real' and r['original_text']]
    print(f"Found {len(text_segments)} segments with real text for training")

    if len(text_segments) >= 10:
        # Prepare training data
        vision_array = np.array([r['vision'] for r in text_segments])
        audio_array = np.array([r['audio'] for r in text_segments])
        text_array = np.array([r['text'] for r in text_segments])

        print(f"\nOriginal shapes:")
        print(f"  Vision: {vision_array.shape}")
        print(f"  Audio: {audio_array.shape}")
        print(f"  Text: {text_array.shape}")

        # Average over frames dimension for vision
        vision_avg = vision_array.mean(axis=1)  # [86, 1024]
        audio_squeezed = audio_array.squeeze(axis=1)  # [86, 1024]
        text_squeezed = text_array.squeeze(axis=1)    # [86, 1024]

        print(f"\nReshaped for training:")
        print(f"  Vision: {vision_avg.shape}")
        print(f"  Audio: {audio_squeezed.shape}")
        print(f"  Text: {text_squeezed.shape}")

        # Convert to tensors
        vision_data = torch.tensor(vision_avg).float().to(device)
        audio_data = torch.tensor(audio_squeezed).float().to(device)
        text_data = torch.tensor(text_squeezed).float().to(device)

        # Initialize trainer
        print("\nCreating IF-MMIN trainer...")
        ifmmin_trainer = IFMMINTrainer(device)

        # Train
        print("\nTraining IF-MMIN...")
        epochs = 30
        losses_history = []

        for epoch in range(epochs):
            losses = ifmmin_trainer.training_step(vision_data, audio_data, text_data)
            losses_history.append(losses)

            if epoch % 5 == 0:
                print(f"Epoch {epoch:2d}: Total Loss = {losses['total_loss']:.4f} "
                      f"(CMD: {losses['cmd_loss']:.4f}, IMG: {losses['img_loss']:.4f})")

        # Final loss
        final_loss = np.mean([l['total_loss'] for l in losses_history[-5:]])
        print(f"\n✓ Training complete! Final avg loss: {final_loss:.4f}")

        # Save model
        model_path = os.path.join(results_dir, "ifmmin_trained.pth")
        ifmmin_trainer.save(model_path)
        print(f"✓ Model saved to: {model_path}")

        # Test on silent segments
        print("\n📝 Testing on silent segments (the ones that got zeros):")
        silent_segments = [r for r in training_results if r['text_source'] == 'zero_fallback']

        if len(silent_segments) > 0:
            for i, seg in enumerate(silent_segments):
                # Average vision frames
                vision_avg = seg['vision'].mean(axis=0)
                vision_tensor = torch.from_numpy(vision_avg).float().to(device).unsqueeze(0)
                audio_tensor = torch.from_numpy(seg['audio'].squeeze()).float().to(device).unsqueeze(0)

                # Imagine text
                imagined_text = ifmmin_trainer.imagine_text(vision_tensor, audio_tensor)
                imagined_emb = imagined_text.cpu().numpy().squeeze()

                # Compare
                zero_energy = np.mean(np.abs(seg['text']))
                imagined_energy = np.mean(np.abs(imagined_emb))

                print(f"\n  Segment {seg['start']:.1f}s-{seg['end']:.1f}s:")
                print(f"    - Zero energy: {zero_energy:.4f}")
                print(f"    - Imagined energy: {imagined_energy:.4f}")
                print(f"    - Ratio: {imagined_energy/(zero_energy+1e-8):.2f}x")
        else:
            print("  No silent segments found in this video")

        # Save training history
        history_path = os.path.join(results_dir, "training_history.npy")
        np.save(history_path, losses_history)
        print(f"✓ Training history saved to: {history_path}")

    else:
        print(f"⚠ Not enough text segments. Need at least 10, have {len(text_segments)}")

else:
    print("⚠ No training data found. Please run Cell 9 first.")
```

```python
# CELL 10.5: Quick verification test

print("="*60)
print("IF-MMIN VERIFICATION TEST")
print("="*60)

# Test on a segment with real text (to see if we can reconstruct)
real_sample = text_segments[0]
vision_tensor = torch.from_numpy(real_sample['vision'].mean(axis=0)).float().to(device).unsqueeze(0)
audio_tensor = torch.from_numpy(real_sample['audio'].squeeze()).float().to(device).unsqueeze(0)
real_text = real_sample['text'].squeeze()

# Imagine text
imagined = ifmmin_trainer.imagine_text(vision_tensor, audio_tensor)
imagined_np = imagined.cpu().numpy().squeeze()

# Compute similarity
real_np = real_text
similarity = np.dot(real_np, imagined_np) / (np.linalg.norm(real_np) * np.linalg.norm(imagined_np))

print(f"\n📊 Reconstruction Test:")
print(f"  Segment time: {real_sample['start']:.1f}s - {real_sample['end']:.1f}s")
print(f"  Original text: \"{real_sample['original_text']}\"")
print(f"  Real embedding energy: {np.mean(np.abs(real_np)):.4f}")
print(f"  Imagined embedding energy: {np.mean(np.abs(imagined_np)):.4f}")
print(f"  Cosine similarity: {similarity:.4f}")

# Test on a silent segment (if any)
if len(silent_segments) > 0:
    silent_sample = silent_segments[0]
    vision_tensor = torch.from_numpy(silent_sample['vision'].mean(axis=0)).float().to(device).unsqueeze(0)
    audio_tensor = torch.from_numpy(silent_sample['audio'].squeeze()).float().to(device).unsqueeze(0)

    imagined_silent = ifmmin_trainer.imagine_text(vision_tensor, audio_tensor)
    imagined_silent_np = imagined_silent.cpu().numpy().squeeze()

    print(f"\n🔇 Silent Segment Test:")
    print(f"  Segment time: {silent_sample['start']:.1f}s - {silent_sample['end']:.1f}s")
    print(f"  Zero embedding energy: {np.mean(np.abs(silent_sample['text'])):.4f}")
    print(f"  Imagined embedding energy: {np.mean(np.abs(imagined_silent_np)):.4f}")
    print(f"  ✓ Now has meaningful embedding (not zeros)!")

print("\n✓ IF-MMIN is ready to use on new videos!")
```

```python
# CELL 11: Test IF-MMIN on a New Video

import os
import torch
import numpy as np
from tqdm import tqdm
from decord import VideoReader, cpu
import matplotlib.pyplot as plt

def test_ifmmin_on_video(video_path, ifmmin_trainer, window_size=2, stride=1):
    """Test IF-MMIN on a new video and compare zero vs imagined embeddings"""

    print(f"\n{'='*60}")
    print(f"TESTING ON: {os.path.basename(video_path)}")
    print(f"{'='*60}")

    # Get segments and transcript
    segments, fps, vr = segment_video(video_path, window_size, stride)
    transcript_segments, full_text = extract_transcript_with_timestamps(video_path, whisper_model)
    segment_tuples = [(None, start, end) for start, end in segments]
    segment_texts = align_text_to_segments(transcript_segments, segment_tuples)

    text_count = sum(1 for t in segment_texts if t)
    print(f"\n📊 Video Stats:")
    print(f"  Total segments: {len(segments)}")
    print(f"  With text: {text_count}")
    print(f"  Without text: {len(segments) - text_count}")

    # Process all segments
    print(f"\n⏳ Processing {len(segments)} segments...")
    results = []

    for idx, (start, end) in enumerate(tqdm(segments, desc="Processing")):
        seg_data = process_segment_with_ifmmin(
            vr, video_path, start, end, fps,
            segment_texts[idx], ifmmin_trainer
        )
        results.append(seg_data)

    # Analyze results
    sources = {}
    for r in results:
        sources[r['text_source']] = sources.get(r['text_source'], 0) + 1

    print(f"\n📊 Text Source Distribution:")
    for source, count in sources.items():
        percentage = (count / len(results)) * 100
        print(f"  {source}: {count} ({percentage:.1f}%)")

    # Compare energies
    real_energies = [np.mean(np.abs(r['text'])) for r in results if r['text_source'] == 'real']
    imagined_energies = [np.mean(np.abs(r['text'])) for r in results if r['text_source'] == 'imagined']

    if real_energies:
        print(f"\n📊 Energy Comparison:")
        print(f"  Real text embeddings: mean energy = {np.mean(real_energies):.4f}")
    if imagined_energies:
        print(f"  Imagined text embeddings: mean energy = {np.mean(imagined_energies):.4f}")
        print(f"  Ratio (imagined/real): {np.mean(imagined_energies)/np.mean(real_energies):.2f}")

    # Show examples of imagined text
    print(f"\n📝 Examples of imagined text embeddings (silent segments):")
    imagined_count = 0
    for r in results:
        if r['text_source'] == 'imagined' and imagined_count < 5:
            energy = np.mean(np.abs(r['text']))
            print(f"  Segment {r['start']:.1f}s-{r['end']:.1f}s: energy = {energy:.4f}")
            imagined_count += 1

    # Visualize timeline
    plt.figure(figsize=(15, 5))

    times = [r['start'] for r in results]
    colors = ['green' if r['text_source'] == 'real' else 'blue' if r['text_source'] == 'imagined' else 'red' for r in results]
    energies = [np.mean(np.abs(r['text'])) for r in results]

    plt.scatter(times, energies, c=colors, alpha=0.6, s=50)
    plt.plot(times, energies, 'gray', alpha=0.3)
    plt.xlabel('Time (seconds)')
    plt.ylabel('Text Embedding Energy')
    plt.title(f'Text Embedding Energy Timeline\n(Green=Real, Blue=Imagined, Red=Zero)')
    plt.grid(True, alpha=0.3)

    # Add legend
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='green', alpha=0.6, label='Real text'),
        Patch(facecolor='blue', alpha=0.6, label='Imagined text (IF-MMIN)'),
        Patch(facecolor='red', alpha=0.6, label='Zero (fallback)')
    ]
    plt.legend(handles=legend_elements)

    plt.tight_layout()
    plt.show()

    return results

# ============================================
# RUN THE TEST ON A NEW VIDEO
# ============================================

# UPDATE THIS PATH to your test video
test_video_path = "/content/drive/MyDrive/PFA/data/test_video.mp4"  # Change this!

if os.path.exists(test_video_path):
    # Run the test
    test_results = test_ifmmin_on_video(
        video_path=test_video_path,
        ifmmin_trainer=ifmmin_trainer,
        window_size=2,
        stride=1
    )

    print(f"\n✅ Test complete! Processed {len(test_results)} segments.")

else:
    print(f"❌ Video not found: {test_video_path}")
    print("Please update the test_video_path to point to your video file.")
```
