```python
import sys
import os
import torch

# 1. Mount Google Drive (if not already done)
from google.colab import drive
drive.mount('/content/drive')

# 2. Configure Path Bridge
PROJECT_ROOT = "/content/drive/MyDrive/PFA"
IMAGEBIND_ROOT = "/content/drive/MyDrive/PFA/ImageBind"
LIB_ROOT ="/content/drive/MyDrive/PFA/ImageBind/ImageBind"

for p in [LIB_ROOT, IMAGEBIND_ROOT, PROJECT_ROOT]:
    if p not in sys.path:
        sys.path.insert(0, p)
        print(f" Linked: {p}")

# 3. Verify folders
INPUT_DIR = os.path.join(PROJECT_ROOT, "data")
OUTPUT_DIR = os.path.join(IMAGEBIND_ROOT, "results")
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"Input: {INPUT_DIR}")
print(f"Output: {OUTPUT_DIR}")
```

```python
from scripts.core import quick_load_all, create_engine
from Text_Handler.preprocessing import TextProducer, run_preprocessing


print("Loading models...")
models = quick_load_all(whisper_size="base")

imagebind_mod = models["imagebind"]
whisper_model = models["whisper"]
blip_processor = models["blip_processor"]
blip_model = models["blip"]
wat_model = models["whisper_at"]
parse_at_label = models["parse_at_label"]
device = models["device"]

print("✓ ImageBind, Whisper, BLIP, and Whisper-AT loaded")

text_producer = TextProducer(
    wat_model=wat_model,
    blip_model=blip_model,
    blip_processor=blip_processor,
    parse_at_label_fn=parse_at_label,
    device=device,
)
print("✓ TextProducer ready")
```

```python
import json

WINDOW_SIZE = 2
STRIDE = 1

segment_data = run_preprocessing(
    VIDEO_PATH,
    text_producer,
    window_size=WINDOW_SIZE,
    stride=STRIDE,
    verbose=True,
)

with open(TEXTS_PATH, "w") as f:
    json.dump(segment_data, f, indent=2)

print(f"✓ Saved {len(segment_data)} segments -> {TEXTS_PATH}")
source_counts = {}
for item in segment_data.values():
    source = item["source"]
    source_counts[source] = source_counts.get(source, 0) + 1
print(f"  Sources breakdown: {source_counts}")
```

```python
engine = create_engine(imagebind_mod, device)
print("✓ ImageBind engine ready")

results = engine.extract_video_features(
    VIDEO_PATH,
    segment_data,
    verbose=True,
)

import numpy as np

np.savez(
    EMB_PATH,
    vision=results["vision"],
    audio=results["audio"],
    text=results["text"],
    times=results["times"],
    raw_text=results["raw_text"],
    text_trust=results["text_trust"],
    errors=np.array(results["errors"], dtype=np.int32),
)

print(f"✓ Saved embeddings -> {EMB_PATH}")
print(f"  Vision shape : {results['vision'].shape}")
print(f"  Audio shape  : {results['audio'].shape}")
print(f"  Text shape   : {results['text'].shape}")
print(f"  Error count  : {len(results['errors'])}")
```

```python
check = np.load(EMB_PATH, allow_pickle=True)
print(f"unified check unavailable here; saved arrays: {list(check.files)}")
print(f"  times shape     : {check['times'].shape}")
print(f"  raw_text count  : {len(check['raw_text'])}")
print(f"  text_trust mean : {check['text_trust'].mean():.3f}")
```

```python
from scripts.core import ConfidenceGate
import torch

V = torch.tensor(results["vision"], dtype=torch.float32)
A = torch.tensor(results["audio"], dtype=torch.float32)
T = torch.tensor(results["text"], dtype=torch.float32)
trust_tensor = torch.tensor(results["text_trust"], dtype=torch.float32)

gate = ConfidenceGate(input_dim=1024, proj_dim=512).eval()

with torch.no_grad():
    unified, weights = gate(V, A, T, trust_tensor)

print(f"Unified: {unified.shape}  norm={unified.norm(dim=-1).mean():.4f}")
for name, w in weights.items():
    print(f"  {name:<8}: mean={w.mean():.3f} ± {w.std():.3f}")

print("\n✓ Stage 4 complete — unified embeddings ready")
```

```python
UNIFIED_PATH = os.path.join(OUTPUT_DIR, f"{VIDEO_STEM}_unified_final.npz")

np.savez(
    UNIFIED_PATH,
    unified=unified.cpu().numpy(),
    times=results["times"],
    raw_text=results["raw_text"],
    text_trust=results["text_trust"],
    weights_vision=weights["vision"].cpu().numpy(),
    weights_audio=weights["audio"].cpu().numpy(),
    weights_text=weights["text"].cpu().numpy(),
)

print(f"✓ Saved unified embeddings -> {UNIFIED_PATH}")

check = np.load(UNIFIED_PATH, allow_pickle=True)
print(f"unified : {check['unified'].shape}")
print(f"times   : {check['times'].shape}")
print(f"Segment 0: t={check['times'][0]}  text='{check['raw_text'][0]}'")
```
