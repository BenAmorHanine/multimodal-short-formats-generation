```python
"""Test that ImageBind is working correctly."""
import torch
import numpy as np
from imagebind.models.imagebind_model import ModalityType
from imagebind import data

print("Running validation test...\n")

try:
    # Test with text embeddings
    text_list = [
        "A dramatic moment",
        "Crowd cheering loudly",
        "A funny scene",
        "An emotional speech"
    ]
    
    inputs = {
        ModalityType.TEXT: data.load_and_transform_text(text_list, device)
    }
    
    with torch.no_grad():
        embeddings = imagebind(inputs)
    
    text_embeddings = embeddings[ModalityType.TEXT]
    
    print(f"вњ“ Text embeddings shape: {text_embeddings.shape}")
    print(f"вњ“ Embedding dimension: {text_embeddings.shape[1]}")
    
    # Compute similarity
    similarity = torch.softmax(text_embeddings @ text_embeddings.T, dim=-1)
    print(f"\nвњ“ Similarity matrix computed")
    print(similarity.cpu().numpy())
    
    print("\nвњ“ ImageBind validation passed!")
    print("вњ“ Setup complete! Ready to extract features.")

except Exception as e:
    print(f"вњ— Validation failed: {e}")
    print("Check that all dependencies are installed correctly.")
```
